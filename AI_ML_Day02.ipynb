```python
import numpy as np

# Create an array
arr = np.array([10, 20, 30, 40])
print("Array:", arr)

# Array operations
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
```

Read / Write CSV Files

```python
# Read dataset
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(data.head())

# Write to CSV
data.to_csv("titanic_copy.csv", index=False)
```

Data Cleaning Operations

```python
# Check missing values
print(data.isnull().sum())

# Fill missing values
data['Age'] = data['Age'].fillna(data['Age'].mean())

# Drop unnecessary columns
data = data.drop(['Cabin', 'Ticket'], axis=1)
```

Filtering & Sorting

```python
# Filter passengers older than 30
adults = data[data['Age'] > 30]

# Sort by Fare
sorted_data = data.sort_values(by='Fare', ascending=False)
```

Aggregation

```python
# Average fare per class
print(data.groupby('Pclass')['Fare'].mean())
```

```python
# Load Titanic dataset
titanic_data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Count missing values
print("Missing values before cleaning:\n", titanic_data.isnull().sum())

# Replace missing Age with mean
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].mean())

# Drop â€œCabinâ€ column
titanic_data = titanic_data.drop('Cabin', axis=1, errors='ignore')

print("\nMissing values after cleaning:\n", titanic_data.isnull().sum())
```

```python
# Filter passengers whose Fare > 50 and Age < 30
filtered_passengers = titanic_data[(titanic_data['Fare'] > 50) & (titanic_data['Age'] < 30)]
print("Passengers with Fare > 50 and Age < 30:\n", filtered_passengers.head())
```

```python
# Group passengers by Pclass and find average age
average_age_by_pclass = titanic_data.groupby('Pclass')['Age'].mean()
print("Average Age by Pclass:\n", average_age_by_pclass)
```
