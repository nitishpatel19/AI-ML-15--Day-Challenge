** 3. Seaborn  Advanced Visualization**

```python
import seaborn as sns
import pandas as pd

# Load sample dataset
data = sns.load_dataset("tips")
print(data.head())

# Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=data)
plt.show()

# Histogram
sns.histplot(data["total_bill"], bins=20, kde=True)
plt.show()

# Boxplot
sns.boxplot(x="day", y="total_bill", data=data)
plt.show()

# Heatmap (correlation matrix)
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

** 4. Mini Project  Visualizing Titanic Dataset**

Goal: Find hidden patterns in passenger survival data.

```python
import seaborn as sns
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# 1. Gender vs Survival
sns.countplot(x='Sex', hue='Survived', data=data)
plt.show()

# 2. Age Distribution
sns.histplot(data['Age'], bins=30, kde=True)
plt.show()

# 3. Survival Rate by Passenger Class
sns.barplot(x='Pclass', y='Survived', data=data)
plt.show()
# 4. Correlation Heatmap
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='Blues')
plt.show()
```

Here's a summary of the important findings from the Titanic dataset visualizations:

**1. Gender vs Survival:** The count plot clearly shows that a significantly higher percentage of females survived compared to males, indicating gender was a strong factor in survival.

**2. Age Distribution:**  The histogram of age distribution suggests that most passengers were young adults, with fewer elderly passengers. The distribution appears right-skewed.

**3. Survival Rate by Passenger Class:** The bar plot demonstrates a clear trend: passengers in higher classes (Pclass 1) had a much higher survival rate than those in lower classes (Pclass 2 and 3), highlighting the impact of socio-economic status on survival.

**4. Correlation Heatmap:** The heatmap provides an overview of correlations between numerical features. Notable correlations include 'Pclass' and 'Survived' (negative correlation, meaning higher class is associated with higher survival), and 'Fare' with 'Pclass' (negative correlation, as expected, higher class pays more).

## Create a Scatter Plot

### Subtask:
Generate a scatter plot to visualize the relationship between two numerical variables from the new dataset, possibly colored by a categorical variable.


**Reasoning**:
I will generate a scatter plot using Seaborn's `scatterplot` function to visualize the relationship between 'sepal_length' and 'sepal_width', differentiating species by color, and adding appropriate titles and axis labels.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Use Seaborn's scatterplot function to create a scatter plot.
# 2. Set 'sepal_length' on the x-axis and 'sepal_width' on the y-axis, using df_iris as the data source.
# 3. Color the points by the 'species' column to differentiate between the different iris species.
plt.figure(figsize=(10, 7))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df_iris, palette='viridis', s=100)

# 4. Add a descriptive title to the plot
plt.title('Sepal Length vs. Sepal Width by Species', fontsize=16)

# 5. Label the x-axis and y-axis clearly
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)

# 6. Display the plot
plt.legend(title='Species')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Scatter Plot Interpretation

The scatter plot of 'Sepal Length vs. Sepal Width by Species' provides several key insights:

*   **Species Separation**: The plot clearly shows that the different `species` of iris flowers (setosa, versicolor, and virginica) form distinct clusters based on their sepal length and sepal width measurements. This indicates that these two features are highly effective in differentiating between the species.
*   **Setosa Characteristics**: The 'setosa' species typically has smaller sepal lengths but tends to have wider sepals compared to the other two species, forming a clearly separated cluster at the bottom left of the plot.
*   **Versicolor and Virginica Overlap**: While 'versicolor' and 'virginica' are somewhat separated, there is a degree of overlap between their sepal length and sepal width ranges, suggesting that these two features alone might not be sufficient to perfectly distinguish them. 'Virginica' generally has larger sepal lengths and widths than 'versicolor'.

This visualization confirms the utility of sepal measurements in botanical classification and highlights the distinctiveness of the 'setosa' species from the others.

## Create a Box Plot

### Subtask:
Generate a box plot to show the distribution of a numerical variable across different categories in the new dataset.


**Reasoning**:
I will generate a box plot using Seaborn's `boxplot` function to visualize the distribution of 'petal_length' across different 'species' in the `df_iris` DataFrame. I will ensure the plot has a descriptive title and clear axis labels for better understanding.

**Reasoning**:
Now, I will generate the Python code to create the box plot as described in the reasoning, using Seaborn's `boxplot` function with 'petal_length' and 'species' from the `df_iris` DataFrame, and include appropriate titles and labels.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Use Seaborn's boxplot function to create a box plot.
# 2. Set 'species' on the x-axis and 'petal_length' on the y-axis, using df_iris as the data source.
plt.figure(figsize=(10, 7))
sns.boxplot(x='species', y='petal_length', data=df_iris, palette='pastel')

# 3. Add a descriptive title to the plot
plt.title('Petal Length Distribution by Species', fontsize=16)

# 4. Label the x-axis and y-axis clearly
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

# 5. Display the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` because `palette` was used without `hue`. To fix this, I will explicitly set `hue='species'` and `legend=False` in the `sns.boxplot` call, as suggested by the warning, to maintain the desired coloring and suppress the warning.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Use Seaborn's boxplot function to create a box plot.
# 2. Set 'species' on the x-axis and 'petal_length' on the y-axis, using df_iris as the data source.
plt.figure(figsize=(10, 7))
sns.boxplot(x='species', y='petal_length', data=df_iris, palette='pastel', hue='species', legend=False)

# 3. Add a descriptive title to the plot
plt.title('Petal Length Distribution by Species', fontsize=16)

# 4. Label the x-axis and y-axis clearly
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

# 5. Display the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Box Plot Interpretation

The box plot titled 'Petal Length Distribution by Species' offers significant insights into the differences in petal lengths among the three iris species:

*   **Clear Separation**: The box plot clearly demonstrates a strong separation in petal lengths among the 'setosa', 'versicolor', and 'virginica' species. This indicates that petal length is a very effective feature for distinguishing between these species.
*   **Setosa Characteristics**: The 'setosa' species has the smallest petal length, with its box and whiskers occupying the lowest range on the y-axis. This species shows very little variability in petal length, as indicated by its small box and short whiskers, suggesting a consistent characteristic.
*   **Versicolor Characteristics**: The 'versicolor' species has intermediate petal lengths, positioned between 'setosa' and 'virginica'. Its distribution is wider than 'setosa', indicating more variability in petal length within this species.
*   **Virginica Characteristics**: The 'virginica' species exhibits the largest petal lengths, with its box and whiskers occupying the highest range. It also shows a moderate degree of variability, similar to or slightly more than 'versicolor'.
*   **No Overlap (Medians)**: There is no overlap in the interquartile ranges (the boxes) of the three species, reinforcing the idea that petal length is a robust distinguishing feature. While some outliers might exist, the central tendency and spread are distinct for each species.

This visualization confirms that petal length is a highly discriminatory feature for classifying iris species, especially in separating 'setosa' from the other two, and also providing good separation between 'versicolor' and 'virginica'.

## Create a Pair Plot

### Subtask:
Generate a pair plot to visualize pairwise relationships between multiple numerical variables in the new dataset, with distributions on the diagonal.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Use Seaborn's pairplot function to create a pair plot.
# 2. Color the points by the 'species' column to differentiate between the different iris species.
# 3. Add a title to the overall plot (though pairplot doesn't have a direct title argument, we can set it for the figure).

g = sns.pairplot(df_iris, hue='species', palette='viridis')
g.fig.suptitle('Pair Plot of Iris Dataset by Species', y=1.02) # y adjusts title position

# 4. Display the plot
plt.show()
```

**Pair Plot Interpretation**
The pair plot provides a comprehensive view of the pairwise relationships and distributions of the numerical features in the Iris dataset, categorized by species:

**Clear Species Separation with Petal Measurements:** The scatter plots involving petal_length and petal_width (e.g., petal_length vs. petal_width, sepal_length vs. petal_length) show excellent separation among the three species. The 'setosa' species forms a very distinct cluster, clearly separated from 'versicolor' and 'virginica'. 'Versicolor' and 'virginica' also show good separation, although with a slight overlap in some areas.

**Sepal Measurements Show More Overlap:** While sepal_length and sepal_width do help in distinguishing species, the scatter plot of sepal_length vs. sepal_width shows more overlap between 'versicolor' and 'virginica' compared to the petal measurements. 'Setosa' remains distinct even with sepal measurements.

**Feature Distributions (Diagonal Histograms):**

**Setosa:** The histograms on the diagonal confirm that 'setosa' has generally smaller petal lengths and widths, and often smaller sepal lengths but broader sepal widths compared to the other species.
**Versicolor and Virginica:** These two species show broader distributions for most features, with 'virginica' generally having larger values than 'versicolor' for sepal_length, petal_length, and petal_width.
**Correlations:** The plots reveal strong positive correlations between petal_length and petal_width, and between sepal_length and petal_length/petal_width, especially within 'versicolor' and 'virginica' groups. This indicates that as one petal (or sepal) dimension increases, the other tends to increase as well, which is characteristic of plant growth.

In summary, the pair plot visually reinforces that petal_length and petal_width are the most discriminative features for classifying Iris species, with sepal_length and sepal_width also contributing, though with less distinct separation for 'versicolor' and 'virginica'.
