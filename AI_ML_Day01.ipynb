```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

if (num1 >= num2) and (num1 >= num3):
   largest = num1
elif (num2 >= num1) and (num2 >= num3):
   largest = num2
else:
   largest = num3

print("The largest number is", largest)
```

```python
num = int(input("Enter a non-negative integer: "))

if num < 0:
    print("Factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    factorial = 1
    for i in range(1, num + 1):
        factorial = factorial * i
    print("The factorial of", num, "is", factorial)
```

```python
import time

count = 10
while count >= 1:
    print(count)
    time.sleep(1) # Pause for 1 second
    count -= 1
print("Blast off!")
```

```python
import random

random_numbers = [random.randint(1, 100) for _ in range(10)] # Generate 10 random numbers between 1 and 100
print("Generated random numbers:", random_numbers)
print("Numbers divisible by 3:")
for num in random_numbers:
    if num % 3 == 0:
        print(num)
```

```python
marks = float(input("Enter the marks: "))

if marks >= 90:
    grade = 'A'
elif marks >= 70:
    grade = 'B'
elif marks >= 50:
    grade = 'C'
else:
    grade = 'F'

print("The grade is:", grade)
```

```python
def calculate_salary(base_salary, experience_years):
    bonus_percentage = 0.10 # 10% bonus
    if experience_years > 3:
        total_salary = base_salary * (1 + bonus_percentage)
        print(f"Experience: {experience_years} years. Bonus applied.")
    else:
        total_salary = base_salary
        print(f"Experience: {experience_years} years. No bonus applied.")
    return total_salary

# Example usage:
print("Salary for 5 years experience:", calculate_salary(50000, 5))
print("Salary for 2 years experience:", calculate_salary(50000, 2))
```
