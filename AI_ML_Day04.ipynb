**‚û§ Bayes‚Äô Theorem (Concept)**

P(A|B) = P(B|A) * P(A) / P(B)

Used in spam detection, medical diagnosis, and Naive Bayes models.

Here's a Python example demonstrating Bayes' Theorem using a hypothetical scenario.

```python
import numpy as np

# Scenario: A medical test for a rare disease
# P(D) = Probability of having the disease (D)
P_D = 0.01  # 1% prevalence

# P(Pos|D) = Probability of testing positive given you have the disease (Sensitivity)
P_Pos_given_D = 0.95 # 95% sensitive

# P(Neg|not D) = Probability of testing negative given you don't have the disease (Specificity)
P_Neg_given_not_D = 0.90 # 90% specific

# P(Pos|not D) = Probability of testing positive given you don't have the disease (False Positive Rate)
P_Pos_given_not_D = 1 - P_Neg_given_not_D # 1 - 0.90 = 0.10

# Calculate P(not D) = Probability of not having the disease
P_not_D = 1 - P_D

# Calculate P(Pos) using the law of total probability:
# P(Pos) = P(Pos|D) * P(D) + P(Pos|not D) * P(not D)
P_Pos = (P_Pos_given_D * P_D) + (P_Pos_given_not_D * P_not_D)

# Apply Bayes' Theorem to find P(D|Pos):
# P(D|Pos) = (P(Pos|D) * P(D)) / P(Pos)
P_D_given_Pos = (P_Pos_given_D * P_D) / P_Pos

print(f"Prevalence of Disease (P(D)): {P_D:.2f}")
print(f"Probability of testing positive given disease (P(Pos|D)): {P_Pos_given_D:.2f}")
print(f"Probability of testing positive given no disease (P(Pos|not D)): {P_Pos_given_not_D:.2f}")
print(f"Probability of testing positive (P(Pos)): {P_Pos:.4f}")
print(f"\nProbability of having the disease given a positive test (P(D|Pos)): {P_D_given_Pos:.4f}")
```

The code successfully calculated the probabilities for a medical test scenario using Bayes' Theorem. Even with a 95% sensitive test and a 1% disease prevalence, the probability of actually having the disease given a positive test is about 8.76%. This highlights how a low prevalence combined with a non-perfect test can lead to a relatively low post-test probability of disease.

2Ô∏è‚É£ Make a 3√ó3 matrix and find its transpose, determinant, inverse

```python
import numpy as np

# Create a 3x3 matrix
matrix = np.array([
    [10, 2, 13],
    [4, 7, 6],
    [7, 23, 9]
])
print("Original Matrix:\n", matrix)

# Find the transpose
transpose_matrix = matrix.T
print("\nTranspose Matrix:\n", transpose_matrix)

# Find the determinant
determinant = np.linalg.det(matrix)
print("\nDeterminant:\n", determinant)

# Find the inverse (if determinant is non-zero)
# A matrix is invertible if and only if its determinant is non-zero.
if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("\nInverse Matrix:\n", inverse_matrix)
else:
    print("\nMatrix is singular, so no inverse exists.")
```

3Ô∏è‚É£ Generate 100 random numbers & calculate:

‚ÄÉ‚Ä¢ Mean, Median, Std Dev, Variance

```python
import numpy as np

# Generate 100 random numbers (e.g., from a uniform distribution between 0 and 100)
random_numbers = np.random.uniform(0, 100, 100)

print("Generated 100 random numbers.")

# Calculate Mean
mean_val = np.mean(random_numbers)
print("Mean:", mean_val)

# Calculate Median
median_val = np.median(random_numbers)
print("Median:", median_val)

# Calculate Standard Deviation
std_dev_val = np.std(random_numbers)
print("Standard Deviation:", std_dev_val)

# Calculate Variance
variance_val = np.var(random_numbers)
print("Variance:", variance_val)
```

4Ô∏è‚É£ Simulate a coin toss 500 times & find probability of getting Heads

```python
import random

# Simulate coin toss 500 times
coin_outcomes = ['Heads', 'Tails']
num_tosses = 500

results = [random.choice(coin_outcomes) for _ in range(num_tosses)]

# Calculate the number of Heads
heads_count = results.count('Heads')

# Calculate the probability of getting Heads
probability_heads = heads_count / num_tosses

print(f"Number of tosses: {num_tosses}")
print(f"Number of Heads: {heads_count}")
print(f"Probability of getting Heads: {probability_heads:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random numbers from a normal distribution
# mean = 0, standard deviation = 1
random_normal_numbers = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram of the generated numbers
plt.hist(random_normal_numbers, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Add a title and labels
plt.title('Normal Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.grid(axis='y', alpha=0.75)
plt.show()
```
